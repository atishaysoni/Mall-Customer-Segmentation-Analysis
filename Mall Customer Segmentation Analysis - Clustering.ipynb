## Clustering

```python
df.head()
```

```python
# Selecting features that are most relevant for clustering analysis
df1 = df[['Annual Income (k$)', 'Spending Score (1-100)']]
df1.head()
```

```python
# scatter plot
sns.scatterplot(x='Annual Income (k$)', y='Spending Score (1-100)', data=df1)
```

```python
# Using the Elbow Method to find the optimal number of clusters
from sklearn.cluster import KMeans
errors = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i)
    kmeans.fit(df1)
    errors.append(kmeans.inertia_)
```

```python
# plot the results for elbow method
plt.figure(figsize=(13,6))
plt.plot(range(1,11), errors)
plt.plot(range(1,11), errors, linewidth=3, color='red', marker='8')
plt.xlabel('No. of clusters')
plt.ylabel('WCSS')
plt.xticks(np.arange(1,11,1))
plt.show()
```

```python
# Applying KMeans with 5 clusters as suggested by the Elbow Method
km = KMeans(n_clusters=5)
km.fit(df1)
y = km.predict(df1)
df1['Label'] = y
df1.head()
```

```python
# Visualizing customer segments using a scatter plot
# Each color represents a distinct cluster
sns.scatterplot(x='Annual Income (k$)', y='Spending Score (1-100)', data=df1, hue='Label', s=50, palette=['red', 'green', 'brown', 'blue', 'orange'])
```

```python
# Visualizing customer segments in 3D
df2 = df[['Annual Income (k$)', 'Spending Score (1-100)', 'Age']]
df2.head()
```

```python
# Using the Elbow Method to find the optimal number of clusters
errors = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i)
    kmeans.fit(df2)
    errors.append(kmeans.inertia_)
```

```python
# plot the results for elbow method
plt.figure(figsize=(13,6))
plt.plot(range(1,11), errors)
plt.plot(range(1,11), errors, linewidth=3, color='red', marker='8')
plt.xlabel('No. of clusters')
plt.ylabel('WCSS')
plt.xticks(np.arange(1,11,1))
plt.show()
```

```python
# Applying KMeans with 5 clusters as suggested by the Elbow Method
km = KMeans(n_clusters=5)
km.fit(df2)
y = km.predict(df2)
df2['Label'] = y
df2.head()
```

```python
# 3d scatter plot
fig = plt.figure(figsize=(20,15))
ax = fig.add_subplot(111, projection='3d')

ax.scatter(df2['Age'][df2['Label']==0], df2['Annual Income (k$)'][df2['Label']==0], df2['Spending Score (1-100)'][df2['Label']==0], c='red', s=50)
ax.scatter(df2['Age'][df2['Label']==1], df2['Annual Income (k$)'][df2['Label']==1], df2['Spending Score (1-100)'][df2['Label']==1], c='green', s=50)
ax.scatter(df2['Age'][df2['Label']==2], df2['Annual Income (k$)'][df2['Label']==2], df2['Spending Score (1-100)'][df2['Label']==2], c='blue', s=50)
ax.scatter(df2['Age'][df2['Label']==3], df2['Annual Income (k$)'][df2['Label']==3], df2['Spending Score (1-100)'][df2['Label']==3], c='brown', s=50)
ax.scatter(df2['Age'][df2['Label']==4], df2['Annual Income (k$)'][df2['Label']==4], df2['Spending Score (1-100)'][df2['Label']==4], c='orange', s=50)
ax.view_init(30, 190)
ax.set_xlabel('Age')
ax.set_ylabel('Annual Income')
ax.set_zlabel('Spending Score')
plt.show()
```

```python
# ## 📌 Conclusion

# This clustering analysis helped identify distinct customer segments based on their annual income and spending behavior. Key observations include:

# - **Cluster 1:** High income, high spending – ideal targets for premium offerings.
# - **Cluster 2:** Low income, low spending – may respond well to discounts and budget products.
# - **Cluster 3:** Moderate income, high spending – great candidates for loyalty programs.
# - **Cluster 4:** High income, low spending – potential to be converted with better marketing.
# - **Cluster 5:** Younger customers with mid-range income and variable spending – good for trend-based campaigns.

# These insights enable more targeted marketing strategies, improve customer satisfaction, and ultimately increase profitability.
```
